```python
# Importing Libraries 
import pandas as pd
```

## Loading in the Data and making into 1NF

```python
# Loading in the data
file_path = '/Users/hari/Big_Data_EPortfolio/DBD_PCOM7E Table.xlsx'
df = pd.read_excel(file_path)
df
```

```python
# Stripping trailing spaces from the column names
df.columns = df.columns.str.strip()
```

```python
structured_data = []

# Filling in the missing values
for i in range(len(df)):
    if pd.notna(df.loc[i, 'Student Number']): 
        student_info = df.loc[i, ['Student Number', 'Student Name', 'Exam Score', 'Support', 'Date of Birth']]
    if pd.notna(df.loc[i, 'Course Name']):
        course_info = df.loc[i, ['Course Name', 'Exam Boards', 'Teacher Name']]
        structured_data.append(pd.concat([student_info, course_info]))

structured_df = pd.DataFrame(structured_data).reset_index(drop=True)
```

```python
# Printing the 1NF dataframe
print(structured_df)
```

## Converting the table into 2NF

```python
# Creating a Students table with student specific information
Students = structured_df[['Student Number', 'Student Name', 'Exam Score', 'Support', 'Date of Birth']].drop_duplicates()
Students.set_index('Student Number', inplace=True)

# Creating a Courses table with course specific information
Courses = structured_df[['Student Number', 'Course Name', 'Exam Boards', 'Teacher Name']]
```

```python
print("Students Table:\n")
print(Students.head())
print("\nCourses Table:\n")
print(Courses.head())
```

## Converting the table into 3NF

```python
# Removing transitive dependencies from the courses table
CourseDetails = Courses[['Course Name', 'Exam Boards', 'Teacher Name']].drop_duplicates()

CourseDetails.set_index('Course Name', inplace=True)

StudentCourses = Courses[['Student Number', 'Course Name']]
```

```python
print("Students Table:\n")
print(Students.head())

print("\nCourse Details Table:\n")
print(CourseDetails)

print("\nStudent Courses Table:\n")
print(StudentCourses)
```
